```python
class Quests:
    def __init__ (self):
        self.quest_info = ""
        self.current_quest = ""
        
    def add_quest(self, giver_name):
        if giver_name == "Commander Cedric1":
            self.quest_info = "Find out who stole from the Humans' stockpile."
            self.current_quest = "Human1"
        elif giver_name == "Torma the Tavernkeep2":
            self.quest_info = "Talk to Tristan."
            self.current_quest = "Human2"
        elif giver_name == "Elf1":
            self.quest_info = "Find out what is killing the plants."
            self.current_quest = "Elf1"
        elif giver_name == "Princess Lyra1":
            self.quest_info = "Interrogate the suspects to find who stole the source of the Princess' magic."
            self.current_quest = "Elf2"
        elif giver_name == "Commander Cedric2":
            self.quest_info = "Get the tree of healing and share it with the Humans."
            self.current_quest = "Human3"
        elif giver_name == "Princess Lyra2":
            self.quest_info = "Get the tree of healing and share it with the Elves."
            self.current_quest = "Elf3"
        
            
    def show_quest(self): #call this if the player presses q
        if 'Human' in self.current_quest or 'Elf' in self.current_quest or "Secret" in self.current_quest:
            print("\nCurrent quest:")
            print(self.quest_info)
        else:
            print("\nYou are not currently on a quest.")
        
    def clear_quest(self): #if the player types 'clear quest'
        print("\nIf you clear this quest, you will automatomatically fail it.")
        print(f"\n{self.quest_info}")
        answer = input(f"Are you sure you want to clear your current quest? Doing so will mean that you cannot do any more quests for the {self.current_quest[:-1]}s. (y/n)").lower()
        while answer != 'y' and answer != 'n':
            answer = input("Please type 'y' or 'n': ").lower()
        if answer == 'y':
            self.remove_quest(self)
    
    def remove_quest(self):
        self.quest_info = ""
        self.current_quest = ""

class Character:
    def __init__(self, description, name, one_line, conversation):
        self.description = description
        self.name = name
        self.one_line = one_line
        self.conversation = conversation
        
    def interact_outside_quest(self):
        print(f"As you pass by, {self.name} says: {self.one_line}\n")
              
    def meet(self):
        print(f"{self.description}\n")
        
    def start_conversation(self, quests_object):
        for char_line, info in self.conversation.items():
            #print(char_line)
            if self.conversation[char_line]["from"] == [""]:
                print(f"{self.name}: {char_line}\n")
                self.choose_dialogue(self.conversation[char_line]["to"], quests_object) #returns initial choices
    
    def choose_dialogue(self, choices, quests_object):
        for num in range(len(choices)):
            print(f"{num+1}. {choices[num]}")#dialogue choices are printed out
        choice = choices[int(input())-1] #player chooses
        print(f"\nYou: {choice}\n")
        if "Quest Obtained" in choice:
            quests_object.add_quest(self.name)
        if choice != "Leave": #if you don't end the conversation
            self.char_responds(choice, quests_object)#character responds
    
    def char_responds(self, player_choice, quests_object):
        for char_line in self.conversation:
            #print(char_line)
            if player_choice in self.conversation[char_line]["from"]: #if what the player chooses triggers this line
                if char_line == "Quest Succeeded":
                    print(f"\n{char_line}")
                    quests_object.remove_quest()
                elif char_line == "Quest Failed":
                    print(f"{char_line}. It seems as if the {quest_object.current_quest[:-1]}s won't be welcoming you back.")
                print(f"{self.name}: {char_line}\n")
                self.choose_dialogue(self.conversation[char_line]["to"], quests_object) #player responds to character




#this conversation dictionary is giant
#here is the basic structure

#{"line which a character can say" : {"from": [lines which the player can say to trigger it], "to":[dialogue choices a player will have after the character says this]}}

#key = string
#value = dictionary of string to list of string pairings

description = "Description"
conversation = {"Hello. Who are you and what is your purpose?":{"from": [""], "to": ["I am [player name] and I am looking for the tree of healing."]},
               "As are we. We have some information on it, but first, you must prove to us that you are worthy.":{"from":["I am [player name] and I am looking for the tree of healing."], "to":["I’m happy to help. What must I do?","I don’t need your approval. I need the information. But I will do it if I have to, so what must I do?", "If you have some information about the tree of healing, why don’t you get it yourself?"]},
               "Because it is far too dangerous":{"from":["If you have some information about the tree of healing, why don’t you get it yourself?"],"to":["I’m happy to help. What must I do?","I don’t need your approval. I need the information. But I will do it if I have to, so what must I do?"]},
               "First you must help us. Someone has been stealing from our stockpile of food and supplies. Find them.":{"from":["I’m happy to help. What must I do?","I don’t need your approval. I need the information. But I will do it if I have to, so what must I do?"],"to":["Quest Obtained. Leave"]}}

# This is the first conversation in human quest 1

my_quest = Quests()
commander = Character(description, "Commander Cedric", "Good day to you, my wayfairing friend.", conversation)
commander.meet()
commander.interact_outside_quest()
commander.start_conversation(my_quest)
```
